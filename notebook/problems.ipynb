# Probability Distribution Problems

**Binomial Distribution**

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either Head or Tails.

```python
# Probability functions, density and distribution

def dbinom(x,size,prob=0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    from scipy.stats import binom
    result=binom.pmf(k=x,n=size,p=prob,loc=0)
    return result

def pbinom(q,size,prob=0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    from scipy.stats import binom
    result=binom.cdf(k=q,n=size,p=prob,loc=0)
    return result

def qbinom(p, size, prob=0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.ppf(q=p,n=size,p=prob,loc=0)
    return result

def rbinom(n,size,prob=0.5):
    """
    Generates random variables from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.rvs(n=size,p=prob,size=n)
    return result
```

```python
dbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) takes the value 2,
pbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) will take a value less than 2
qbinom(0.9, size = 10, prob = 0.2) # what value of a binomial (10,0.2) has a cumulative probability of 0.9?
rbinom(2000, size = 10, prob = 0.2) # generate 2000 random values from a binomial distribution (10,0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Discrete distributions
xs = np.arange( 1, 11 )
ys = dbinom(xs,size=10,prob=0.5) # compute the shape of the distribution
plt.plot( xs, ys, 'o' )   # plot circles...
plt.vlines( xs, 0, ys )   # ...and lines
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function binomial distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, binom

# Probability functions, density, and distribution for binomial distribution
def dbinom(x, size, prob=0.5):
    result = binom.pmf(k=x, n=size, p=prob, loc=0)
    return result

def pbinom(q, size, prob=0.5):
    result = binom.cdf(k=q, n=size, p=prob, loc=0)
    return result

def qbinom(p, size, prob=0.5):
    result = binom.ppf(q=p, n=size, p=prob, loc=0)
    return result

def rbinom(n, size, prob=0.5):
    result = binom.rvs(n=size, p=prob, size=n)
    return result

# Probability functions, density, and distribution for normal distribution
def dnorm(x, mean=0, std_dev=1):
    result = norm.pdf(x, loc=mean, scale=std_dev)
    return result

def pnorm(q, mean=0, std_dev=1):
    result = norm.cdf(q, loc=mean, scale=std_dev)
    return result

def qnorm(p, mean=0, std_dev=1):
    result = norm.ppf(p, loc=mean, scale=std_dev)
    return result

def rnorm(n, mean=0, std_dev=1):
    result = norm.rvs(loc=mean, scale=std_dev, size=n)
    return result

# Probability functions, density, and distribution for uniform distribution
def dunif(x, min_val=0, max_val=1):
    result = uniform.pdf(x, loc=min_val, scale=max_val-min_val)
    return result

def punif(q, min_val=0, max_val=1):
    result = uniform.cdf(q, loc=min_val, scale=max_val-min_val)
    return result

def qunif(p, min_val=0, max_val=1):
    result = uniform.ppf(p, loc=min_val, scale=max_val-min_val)
    return result

def runif(n, min_val=0, max_val=1):
    result = uniform.rvs(loc=min_val, scale=max_val-min_val, size=n)
    return result

# Plotting binomial distribution
xs_binom = np.arange(1, 11)
ys_binom = dbinom(xs_binom, size=10, prob=0.5)
plt.subplot(131)
plt.plot(xs_binom, ys_binom, 'o')
plt.vlines(xs_binom, 0, ys_binom)
plt.ylim(bottom=0)
plt.title('Binomial Distribution')

# Plotting normal distribution
xs_norm = np.linspace(-5, 5, 100)
ys_norm = dnorm(xs_norm, mean=0, std_dev=1)
plt.subplot(132)
plt.plot(xs_norm, ys_norm)
plt.title('Normal Distribution')

# Plotting uniform distribution
xs_unif = np.linspace(0, 1, 100)
ys_unif = dunif(xs_unif, min_val=0, max_val=1)
plt.subplot(133)
plt.plot(xs_unif, ys_unif)
plt.title('Uniform Distribution')

plt.tight_layout()
plt.show()
```
